```python
# # Label Encoding
```

```python
# Have to remap the class labels for convenience, no liver disease then:=0 for patients having liver disease then:=1
```

```python
liver_df['Dataset'] = liver_df['Dataset'].map({2:0,1:1})
```

```python
liver_df['Dataset'].value_counts()
```

```python
# # Encoding -2  Male and Female
```

```python
def binary_encode(df, column, positive_value):
    df = df.copy()
    df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df
```

```python
liver_df = binary_encode(liver_df, 'Gender', 'Male')
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import graphviz
split_data = train_test_split(liver_df, test_size=.20)

train_data = split_data[0]
test_data = split_data[1]

Dataset_train = train_data.Dataset.tolist()
Dataset_test = test_data.Dataset.tolist()

train_data_remove = train_data.drop(columns = "Dataset")
test_data_remove = test_data.drop(columns = "Dataset")
```

```python
import sklearn
clf2 = sklearn.tree.DecisionTreeClassifier()

clf2.fit(train_data_remove, Dataset_train)

dot_data = tree.export_graphviz(clf2, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("liver_df")

graph
```

```python
test_array2 = clf2.predict(test_data_remove).tolist()
CTestAccuracy = sklearn.metrics.accuracy_score(Dataset_test, test_array2)

print('C4.5 Testing accuracy: ',CTestAccuracy)
```
